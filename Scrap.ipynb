```python
from bs4 import BeautifulSoup as BS
import pandas as pd
import requests 
```

```python
url="https://developers.google.com/machine-learning"
```

```python
page = requests.get(url)
print(page.status_code)
```

```python
soup = BS(page.content ,  "html.parser")
```

```python
categores = soup.find_all("figure", class_= "devsite-landing-row-item-image")
categores
```

```python
links = []

for category in categores :
    for a in category.find_all("a", href=True):
        links.append(a["href"])

print(links)
```

```python
links2=[]
for link in links : 
    page = requests.get(link)
    soup = BS(page.content, "html.parser")
    content = soup.find("div", class_="devsite-mobile-nav-bottom")

    if content:
        for a in content.find_all("a", href=True):
            links2.append(a["href"])

links2
```

```python
links.pop(1) 
```

```python
content = soup.find("div", class_="devsite-article-body clearfix")
```

```python
data = {
    "page_links" : [] , 
    "text" : []
    }

for link in links :
    page = requests.get(link)
    soup = BS(page.content ,  "html.parser")
    content = soup.find("div", class_= "devsite-article-body clearfix")
    if content:
     for p in content.find_all("p"):
        text = p.get_text(separator=" ", strip=True)
        data["page_links"].append(link)
        data["text"].append(text)
```

```python
for link in links2 :
    page = requests.get("https://developers.google.com/" + link)
    soup = BS(page.content ,  "html.parser")
    content = soup.find("div", class_= "devsite-article-body clearfix")
    if content:
     for p in content.find_all("p"):
        text = p.get_text(separator=" ", strip=True)
        data["page_links"].append(link)
        data["text"].append(text)
```
